```python
def make_discrete(imdb_score):
    if imdb_score >= 8:
        return 'iyi bir film'
    elif (imdb_score>4) and (imdb_score < 8):
        return 'ortalama bir film'
    else:
        return 'kötü bir film'

movie['imdb_score_grup'] = movie['imdb_score'].apply(make_discrete)
display(movie.head())
movie.groupby("imdb_score_grup")["imdb_score"].count()
```
